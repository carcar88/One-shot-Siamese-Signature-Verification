```python
import os
import shutil
from tqdm import tqdm
```

```python
SOURCE_FORG_DIR = "full_forg"
SOURCE_ORG_DIR = "full_org"
TRAIN_DIR = "train"
TEST_DIR = "test"
```

```python
def process_folder(source_path, is_forgery):
    files = [f for f in os.listdir(source_path)]

    for filename in tqdm(files):
        # split berdasarkan underscore
        parts = filename.split('_')

        # buat path untuk file
        if len(parts) < 3:
            continue
        
        writer_id = parts[1]
        
        if 1 <= int(writer_id) <= 45:
            target_dir = TRAIN_DIR
        else:
            target_dir = TEST_DIR

        writer_folder_name = f"writer_{writer_id}"
        type_folder_name = "forged" if is_forgery else "genuine"

        final_folder_path = os.path.join(target_dir, writer_folder_name, type_folder_name)

        os.makedirs(final_folder_path, exist_ok=True)

        # copy file ke folder baru
        src_file = os.path.join(source_path, filename)
        dst_file = os.path.join(final_folder_path, filename)
        shutil.copy2(src_file, dst_file)
    
```

```python
def organize_signatures():
    # buat folder kalau foldernya belum ada
    if not os.path.exists(TRAIN_DIR):
        os.makedirs(TRAIN_DIR)
        
    if not os.path.exists(TEST_DIR):
        os.makedirs(TEST_DIR)

    # proses untuk original dataset
    if os.path.exists(SOURCE_ORG_DIR):
        process_folder(SOURCE_ORG_DIR, is_forgery=False)
    else:
        print(f"Error: Original folder {SOURCE_ORG_DIR} tidak ditemukan")

    # proses forged dataset
    if os.path.exists(SOURCE_FORG_DIR):
        process_folder(SOURCE_FORG_DIR, is_forgery=True)
    else:
        print(f"Error: Forged folder {SOURCE_FORG_DIR} tidak ditemukan")
```

```python
organize_signatures()
```
